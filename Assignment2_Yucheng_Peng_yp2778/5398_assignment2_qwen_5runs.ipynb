```python
from google.colab import drive
import huggingface_hub
from google.colab import userdata
import pandas as pd
import json
import os
import torch
import time
```

```python
from tqdm import tqdm
```

```python
df_test = pd.read_json("/content/drive/MyDrive/5398/LLaMA-Factory/data/df_test_system.json")

results = []

for idx, row in tqdm(df_test.iterrows(), total=len(df_test)):
    system_text = row["system"] if pd.notna(row["system"]) else ""

    user_content = row["instruction"]
    if pd.notna(row.get("input", "")) and row.get("input", "") != "":
      user_content += "\n" + row["input"]

    full_prompt = system_text + "\n" + user_content

    messages = [
        {"role": "user", "content": full_prompt}
    ]

    # Run the model 5 times for the same input
    for run_num in range(5):
        start_time = time.time()

        response = ""
        for new_text in chat_model.stream_chat(messages):
            response += new_text

        output = response

        inference_time = time.time() - start_time

        results.append({
            "id": idx, # Use index as the unique id for the same input
            "run_num": run_num + 1,
            "prompt": row["instruction"],
            "system": row.get("system", ""),
            "reference": row["output"],
            "prediction": output,
            "inference_time_s": inference_time
        })

        # Optional: Call torch_gc() if it's defined elsewhere, assuming it's for clearing CUDA cache
        # torch_gc()

df_results = pd.DataFrame(results)
df_results.to_csv("/content/drive/MyDrive/5398/LLaMA-Factory/Qwen2.5_test_results_5runs.csv", index=False)
```

# 5runs results

normal vote

```python
import pandas as pd
```

```python
df_ini = pd.read_csv("/content/drive/MyDrive/5398/LLaMA-Factory/Qwen2.5_test_results_5runs.csv")
```

```python
def extract_text(s):
    if pd.isna(s):
        return None
    import re
    match = re.search(r'\[Prediction & Analysis\]:(.*?)Analysis:', s, re.S)
    return match.group(1).strip() if match else None


df_ini['real'] = df_ini['reference'].apply(extract_text)
df_ini['pred'] = df_ini['prediction'].apply(extract_text)
```

```python
df_ini.head()
```

```python
df_ini.shape
```

```python
def get_label(s):
    if pd.isna(s):
        return None
    s = s.lower()
    if 'up' in s:
        return 'up'
    elif 'down' in s:
        return 'down'
    return None

df_ini['real_label'] = df_ini['real'].apply(get_label)
df_ini['pred_label'] = df_ini['pred'].apply(get_label)

df_ini['group'] = df_ini.index // 5

def vote(series):
    counts = series.value_counts()
    if len(counts) == 0:
        return None
    return counts.idxmax()

df_grouped = df_ini.groupby('group').agg({
    'real_label': 'first',
    'pred_label': vote
}).reset_index()

df_grouped['result'] = (df_grouped['real_label'] == df_grouped['pred_label']).astype(int)

df_5run_results = df_grouped[['real_label', 'pred_label', 'result']]
```

```python
df_5run_results
```

```python
valid_mask = df_5run_results['real_label'].notna() & df_5run_results['pred_label'].notna()

y_true = df_5run_results.loc[valid_mask, 'real_label']
y_pred = df_5run_results.loc[valid_mask, 'pred_label']
```

```python
from sklearn.metrics import confusion_matrix

labels = ['up', 'down']
cm = confusion_matrix(y_true, y_pred, labels=labels)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.imshow(cm)
plt.title("Confusion Matrix")
plt.colorbar()

plt.xticks(range(len(labels)), labels)
plt.yticks(range(len(labels)), labels)

plt.xlabel("Predicted")
plt.ylabel("Actual")

for i in range(len(labels)):
    for j in range(len(labels)):
        plt.text(j, i, cm[i, j], ha='center', va='center')

plt.tight_layout()
plt.show()
```

```python
def get_label(s):
    if pd.isna(s):
        return None
    s = s.lower()
    if 'up' in s:
        return 'up'
    elif 'down' in s:
        return 'down'
    return None

df_ini['real_label'] = df_ini['real'].apply(get_label)
df_ini['pred_label'] = df_ini['pred'].apply(get_label)

df_ini['group'] = df_ini.index // 5

def vote(series):
    series = series.dropna()
    if len(series) == 0:
        return None

    down_count = (series == 'down').sum()

    if down_count >= 2:
        return 'down'
    else:
        return 'up'

df_grouped = df_ini.groupby('group').agg({
    'real_label': 'first',
    'pred_label': vote
}).reset_index()

df_grouped['result'] = (df_grouped['real_label'] == df_grouped['pred_label']).astype(int)

df_5run_results = df_grouped[['real_label', 'pred_label', 'result']]
```

```python
from sklearn.metrics import confusion_matrix

labels = ['up', 'down']
cm = confusion_matrix(y_true, y_pred, labels=labels)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.imshow(cm)
plt.title("Confusion Matrix")
plt.colorbar()

plt.xticks(range(len(labels)), labels)
plt.yticks(range(len(labels)), labels)

plt.xlabel("Predicted")
plt.ylabel("Actual")

for i in range(len(labels)):
    for j in range(len(labels)):
        plt.text(j, i, cm[i, j], ha='center', va='center')

plt.tight_layout()
plt.show()
```
