```python
def run_one_row(system_text, company_text, news_text):
    response = ""

    messages = [
        {
            "role": "user",
            "content": system_text + "\n\n" + company_text + "\n\n" +news_text
        }
    ]

    for new_text in chat_model.stream_chat(messages):
        response += new_text

    return response
```

```python
from tqdm import tqdm

tqdm.pandas()

df_exploded['output'] = df_exploded.progress_apply(
    lambda row: run_one_row(row['system'], row['company_intro'], row['new']),
    axis=1
)
```

```python
import json
import re
import pandas as pd
def extract_json(text):
    if pd.isna(text):
        return None

    text = str(text)

    match = re.search(r"\{.*\}", text, re.DOTALL)
    if not match:
        return None

    json_str = match.group()

    try:
        return json.loads(json_str)
    except:
        return None
```

```python
df = pd.read_csv("/content/drive/MyDrive/5398/LLaMA-Factory/df_news_analyst_qwen_output.csv")
```

```python
df.head()
```

```python
df["parsed"] = df["output"].apply(extract_json)

df = df[df["parsed"].notnull()].copy()
```

```python
#result is not stable enough, a small precentage of data doesn't provide"urgency" result
df = df[df["parsed"].apply(lambda x: isinstance(x, dict) and "urgency" in x)].copy()

df["sentiment"] = df["parsed"].apply(lambda x: x["sentiment"])
df["urgency"] = df["parsed"].apply(lambda x: x["urgency"])
df["relevance"] = df["parsed"].apply(lambda x: x["relevance"])
df["confidence"] = df["parsed"].apply(lambda x: x["confidence"])
df["event_type"] = df["parsed"].apply(lambda x: x["event_type"])
```

```python
df["signal"] = (
    df["sentiment"] *
    df["urgency"] *
    df["relevance"] *
    df["confidence"]
)
```

```python
weekly_signal = df.groupby("id").apply(
    lambda x: (
        (x["sentiment"] * x["urgency"] * x["relevance"] * x["confidence"])
        .sum()
    )
)
```

```python
weekly_signal = (weekly_signal - weekly_signal.mean()) / weekly_signal.std()
```

```python
weekly_signal.head()
```

```python
def to_position(x):
    if x > 1:
        return 1
    elif x < -1:
        return -1
    else:
        return 0

positions = weekly_signal.apply(to_position)
```
